```python
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure job_posted_date is a datetime object
full_dedup['job_posted_date'] = pd.to_datetime(full_dedup['job_posted_date'])

# Set job_posted_date as the index
full_dedup.set_index('job_posted_date', inplace=True)

# Group data by week and job_title, then count occurrences
grouped_data = (
    full_dedup
    .groupby([pd.Grouper(freq='W'), 'job_title'])
    .size()
    .reset_index(name='count')
)

# Plot the line chart with Seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=grouped_data, x='job_posted_date', y='count', hue='job_title', marker='o')

plt.xticks(rotation=45)
plt.title('Weekly Job Posted Count (Hue: Job Title)')
plt.xlabel('Week')
plt.ylabel('Count')
plt.legend(title='Job Title', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
